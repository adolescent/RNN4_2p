```python
import torch
from torchvision import datasets
from torchvision.transforms import ToTensor, Lambda
import matplotlib.pyplot as plt

# for many network label should be given as onehot.
ds = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,
    transform=ToTensor(),
    target_transform=Lambda(lambda y: torch.zeros(10, dtype=torch.float).scatter_(0, torch.tensor(y), value=1))# this change index into one hot format.
)
```

```python
# show data.
labels_map = {
    0: "T-Shirt",
    1: "Trouser",
    2: "Pullover",
    3: "Dress",
    4: "Coat",
    5: "Sandal",
    6: "Shirt",
    7: "Sneaker",
    8: "Bag",
    9: "Ankle Boot",
}# label number with given name.

sample_idx = torch.randint(len(ds), size=(1,)).item() # random show 1 items.
img, label = ds[sample_idx]
plt.imshow(img.squeeze(), cmap="gray")
```
